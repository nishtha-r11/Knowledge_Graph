```python
import pandas as pd
from transformers import pipeline
```

```python
df = pd.read_csv("wikidata_properties_labels_sorted.csv")
# df = subset_iloc = df.iloc[:30]
# df = subset_iloc = df.iloc[:300]
df
```

```python
classifier = pipeline("zero-shot-classification", model = "facebook/bart-large-mnli")
clabels = ["movies", "TV programs", "sports programs", "news programs"]

cthreshold = 0.5

# def category(description):
#     res = classifier(description, clabels)
#     plabels = res["labels"][0]
#     cscore = res["scores"][0]
#     if cscore < cthreshold:
#         return "negative"
#     return plabels

def category(description):
    if "identifier" in description or "ID" in description:
        return "negative"
    res = classifier(description, clabels)
    plabels = res["labels"][0]
    cscore = res["scores"][0]
    if cscore < cthreshold:
        return "negative"
    return plabels

df["predicted_labels"] = df["Description"].apply(category)
df
```
